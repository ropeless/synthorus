# Developer Community

Synthorus is a community-driven open source project. You are very welcome to join the Synthorus developer community.
If you want to join in, please read the [CONTRIBUTING.md](https://github.com/ropeless/synthorus/blob/main/CONTRIBUTING.md) guidelines
and the [CODE_OF_CONDUCT.md](https://github.com/ropeless/synthorus/blob/main/CODE_OF_CONDUCT.md).

Software and documentation issues may be raised using the GitHub issue tracker, [github.com/ropeless/synthorus/issues](https://github.com/ropeless/synthorus/issues).


## Developer Installation

A developer installation will normally clone the origin repository. The project source code is available at https://github.com/ropeless/synthorus.

The project uses [poetry](https://python-poetry.org/docs/) to manage Python dependencies.

To get started...

1. Clone the repository, e.g., using `git clone https://github.com/ropeless/synthorus.git` or `git clone git@github.com:ropeless/synthorus.git`.
2. Install Python dependencies using poetry: `poetry sync`.
   If poetry fails you may need to first run `python -m ensurepip --upgrade` for some versions of Python, to ensure `pip` is operational.

A Python script called `synthorus_demos/all_demos.py` is available to run most demo scripts in Synthorus (with some exclusions for practical reasons). This can be used as a smoke test.

A Python script called `tests/all_test.py` is available to run all unit tests in Synthorus. The script `tests/all_coverage.py` is available to run all unit tests with code coverage analysis.
Unit tests are in the package `tests` and are named _something_`_test.py`, where _something_ typically is a direct reference to a `synthorus` module.

To build a copy of the Jupyter Book documentation, run the Python script `build_docs.py`.
