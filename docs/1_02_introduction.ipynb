# Introduction

## What is Synthorus

Synthorus is a software library for creating synthetic data.

The library is provided as a Python package (called `synthorus`). There is an additional package containing demonstration Python scripts (called `synthorus_demos`).

### Aim

The aims of Synthorus are:
  - to provide a Python library for building synthetic data simulators
    based on probabilistic graphical models
  - to be efficient, flexible, and easy to use
  - to exhibit excellent design, code, and documentation
  - to support researchers and businesses wanting to explore and use
    synthetic data.


### Key features

- unlimited synthetic data from a Synthorus simulator
- can produce synthetic data for relational data models (i.e., multiple data entities), with guaranteed referential integrity
- integrated privacy protection using Differential Privacy and cryptographically secure random number generator
- integrated accuracy evaluation of synthetic data for verifying synthetic data utility
- based on probabilistic graphical models for mathematically verifiable statistical properties
- coded in pure Python
- fast, flexible and efficient.


### How it works

Synthorus lets you build a simulator that generates synthetic data **records** for one or more data **entities**.
A data entity has a set of **fields** and each data record of an entity will have a **value** for each field of the entity.

Associated with each entity is a **probabilistic graphical model** (PGM).
A PGM defines a joint probability distribution over a set of **random variables**.
Values for records are produced by sampling random variables or a computing a function of other fields.
Using PGMs for random data allows Synthorus to make strong guarantees about the statistics of synthetic data.

**Entity relationships** are implemented using **primary keys** and **foreign keys**, as is typical for databases.
Synthorus ensures that values for primary keys are generated before being referenced by a foreign key. This guarantees
the referential integrity of synthetic data.

The particular number of records holding a foreign key to the primary key of a specific record is called the **record cardinality**.
Synthorus provides a flexible and efficient mechanism to control record cardinality that is easy to understand and specify.
Record cardinality can be specified based on any combination of fixed counts, random variables, field comparisons, and testing of values.

When a PGM is sampled to produce a synthetic value, the probabilities can be conditioned on other fields, including
fields from related entities and previously generated records. That is, Synthorus lets you create arbitrary hierarchical Markov models.

PGMs can be manually specified, but the expectation is that they will normally be learned from reference data. Synthorus provides
many ways to specify **data sources** for reference data. These include direct specification of data, reading many file formats, and direct database access.

When reference data is sensitive, Synthorus provides **Differential Privacy** to protect against leaking sensitive data and
reidentification attacks.
In essence, noise is added to PGM parameters creating a tradeoff across privacy protection and data accuracy. This noise
does not affect referential integrity or field validity.
Synthorus uses **cryptographically secure random number generation** to protect against attacks against sampling.
Synthorus provides fine-grained control over how Differential Privacy is applied.

When a Synthorus simulator is running it emits synthetic data records.
Synthorus provides a general system of **sim recorders** that can write records as they are emitted.
Standard sim recorders are provided for debugging, CSV, Pandas dataframe, and JSON formats. This
system is easy to customize for arbitrary requirements.

Synthorus provides integrated reporting tools to report on privacy protection and data accuracy. As PGMs are used for the source
of random values, the PGMs are directly queried for assessing accuracy. This improves the reliability of accuracy reports and
avoids the need to generate synthetic records just to report on system performance.

Importantly, privacy protection and data accuracy guarantees are made on a Synthorus simulator, rather than synthetic data directly.
Privacy protection guarantees on the synthetic data are inherited from the simulator by virtue of Differential Privacy (via the "post-processing" property).
Data accuracy guarantees on the synthetic data are inherited from the simulator by virtue of a PGM mathematically defining a joint probability distribution.
This means that once a Synthorus simulator is created and verified, it can produce **unlimited synthetic data** without concern about privacy or accuracy.
