# Installation

## Quick Start

Synthorus is hosted on [PyPI](https://pypi.org/project/synthorus/) so can be
installed into your Python environment using `pip` or other Python package manager, e.g.,
```console
pip install synthorus
```

You can then create and run synthic data simulations.

Here is a demo simulation that writes simulated records to stdout.

````python
from synthorus.simulator.sim_recorder import DebugRecorder
from synthorus_demos.simulator import demo_simulator_1_entity
from synthorus.simulator.simulator import Simulator

simulator: Simulator = demo_simulator_1_entity.make_simulator()

simulator.run(DebugRecorder())
````

You can run the standard demo scripts as a check of the installation using the following Python script.
```python
from synthorus_demos.all_demos import main
main()
```
