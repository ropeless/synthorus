# Introduction

## What is Synthorus

Synthorus is a software library for creating synthetic data.

The library is provided as a Python package (called `synthorus`). There is an additional package containing demonstration Python scripts (called `synthorus_demos`).

### Aim

The aims of Synthorus are:
  - to provide a Python library for building synthetic data simulators
    based on probabilistic graphical models
  - to be efficient, flexible, and easy to use
  - to exhibit excellent design, code, and documentation
  - to support researchers and businesses wanting to explore and use
    synthetic data.


### Key features

- unlimited synthetic data from a Synthorus simulator
- can produce synthetic data for relational data models (i.e., multiple data entities), with guaranteed referential integrity
- integrated privacy protection using Differential Privacy and cryptographically secure random number generator
- integrated accuracy evaluation of synthetic data for verifying synthetic data utility
- based on probabilistic graphical models for mathematically verifiable statistical properties
- coded in pure Python
- fast, flexible and efficient.
